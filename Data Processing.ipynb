```
features = features.append(test_data_2.iloc[33416:,:]) #All data from training, validating and testing set
features = features.reset_index()
```

```
#MACD 
SMA_12 = features.rolling(window=12)['Weighted Volume Price'].mean()
multiplier_12 = 2/13
EMA_12 = multiplier_12*features['Weighted Volume Price']+(1-multiplier_12)*SMA_12.shift()

SMA_26 = features.rolling(window=26)['Weighted Volume Price'].mean()
multiplier_26 = 2/27
EMA_26 = multiplier_26*features['Weighted Volume Price']+(1-multiplier_26)*SMA_26.shift()

MACD = EMA_12-EMA_26
features['MACD'] = MACD
```

```
#1-7 hourly return
features['1 hour return'] = features['Weighted Volume Price'].pct_change(1)
features['2 hour return'] = features['Weighted Volume Price'].pct_change(2)
features['3 hour return'] = features['Weighted Volume Price'].pct_change(3)
features['4 hour return'] = features['Weighted Volume Price'].pct_change(4)
features['5 hour return'] = features['Weighted Volume Price'].pct_change(5)
features['6 hour return'] = features['Weighted Volume Price'].pct_change(6)
features['7 hour return'] = features['Weighted Volume Price'].pct_change(7)
```

```
#Moving average
features['50 hour moving average'] = features.rolling(window=50)['Weighted Volume Price'].mean()
features['200 hour moving average'] = features.rolling(window=200)['Weighted Volume Price'].mean()
```

```
#Rolling volatility
features['20 hour volatility'] = features['Weighted Volume Price'].rolling(window=20).std()
features['100 hour volatility'] = features['Weighted Volume Price'].rolling(window=100).std()
features['200 hour volatility'] = features['Weighted Volume Price'].rolling(window=200).std()
```

```
#RSI
def divbyzero(x,y): #create this function in the case where there is no negative return in 14 days period.
    if x == 0:
        a=0
    elif y ==0:
        a = 100
    else:
        a= x/y
    return a
RSI = np.zeros(features['Weighted Volume Price'].shape[0])

for x in range(RSI.shape[0]):
    if x <13:
        continue
    else:
        pos = features['Weighted Volume Price'][(x-14):x][features['1 hour return'][(x-14):x]>0].mean()
        neg = features['Weighted Volume Price'][(x-14):x][features['1 hour return'][(x-14):x]<0].mean()
        if np.isnan(pos)==1:
            pos = 0
        if np.isnan(neg)==1:
            neg = 0
                 
        rs = divbyzero(pos,neg)
        RSI[x]=100-100/(1+rs)

features['RSI'] = RSI
features
```

```
## Saving the dataframe
features.to_csv("/data/workspace_files/features_draft1.csv")
```

```
## Feature creation - bollinger band
## Currently creating a volatility band on 'n' hours
def get_bollinger_bands(prices, window_size):
    sma = prices.rolling(window_size).mean()
    std = prices.rolling(window_size).std()
    bollinger_up = sma + std*2
    bollinger_down = sma - std*2
    return bollinger_up, bollinger_down

temp_prices = features["Weighted Volume Price"]
features["bollinger_up_7hr"], features["bollinger_down_7hr"] = get_bollinger_bands(temp_prices, window_size=7)
```

```
## Create a boolean feature, indicating breach of bollinger bands
def bollinger_breach(cell):
    if cell["Weighted Volume Price"] > cell["bollinger_up_7hr"]:
        return 1
    elif cell["Weighted Volume Price"] < cell["bollinger_down_7hr"]:
        return -1
    else:
        return 0
features["bollinger_breach"] = features.apply(bollinger_breach, axis=1)
```

```
## Create stochastic oscillator indicator
def stochastic_oscillator(prices, window=14):
    high = prices.rolling(window).max()
    low = prices.rolling(window).min()
    # if high == low:
    #     stoch = 0
    # else:
    numerator = 100*(prices - low)
    denominator = (high - low)
    stoch = np.zeros(numerator.shape)
    for i in range(numerator.shape[0]):
        if denominator[i] != 0:
            stoch[i] = numerator[i]/ denominator[i]
        else:
            stoch[i] = 0
    return stoch

stoch = stochastic_oscillator(temp_prices)
```

```
features["stoch"] = stoch
```

```
# feature selection
from sklearn.feature_selection import mutual_info_regression
X_data = features.set_index('Date')
cols = ['Weighted Volume Price', 'Total Volume', 'MACD',
       '1 hour return', '2 hour return', '3 hour return', '4 hour return',
       '5 hour return', '6 hour return', '7 hour return',
       '50 hour moving average', '200 hour moving average',
       '20 hour volatility', '100 hour volatility', '200 hour volatility',
       'RSI', 'bollinger_up_7hr', 'bollinger_down_7hr', 'bollinger_breach',
       'stoch']
X_data = X_data[199:]
X_data = X_data[cols]
target = (X_data['Weighted Volume Price'].shift(-7)/X_data['Weighted Volume Price'] - 1).iloc[:-7]
X_data = X_data.iloc[:-7]
assert (target.index == X_data.index).all()

threshold = 10
feature_scores = mutual_info_regression(X_data, target, random_state=0)
high_score_features = []
for score, f_name in sorted(zip(feature_scores, X_data.columns), reverse=True)[:threshold]:
    print(f_name, score)
    high_score_features.append(f_name)

high_score_features
```

```
high_score_features.append('Date')
ten_feature_set = features[high_score_features]
ten_feature_set.to_csv('/data/workspace_files/ten_feature_set.csv', index=False)
```
